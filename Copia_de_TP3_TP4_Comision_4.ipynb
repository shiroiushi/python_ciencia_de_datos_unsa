##***EJERCICIO 1***

Simular el lanzamiento de un dado eqilibrado de 6 caras.


*   Mostrar la probabilida aproximada de obtener el número 2, 3 y 5
*   Graficar el histograma de las probabilidades



```python
import numpy as np
import matplotlib.pyplot as plt

# Pedir al usuario el número de lanzamientos a simular
num_lanzamientos = int(input("Ingrese el número de lanzamientos a simular: "))

# Simular el lanzamiento de un dado el número de veces indicado
lanzamientos = np.random.randint(1, 7, size=num_lanzamientos)

# Calcular las probabilidades aproximadas para el 2, 3 y 5
probabilidad_2 = np.sum(lanzamientos == 2) / len(lanzamientos)
probabilidad_3 = np.sum(lanzamientos == 3) / len(lanzamientos)
probabilidad_5 = np.sum(lanzamientos == 5) / len(lanzamientos)

print(f"Probabilidad aproximada de obtener el número 2: {probabilidad_2:.2f}")
print(f"Probabilidad aproximada de obtener el número 3: {probabilidad_3:.2f}")
print(f"Probabilidad aproximada de obtener el número 5: {probabilidad_5:.2f}")

# Crear el histograma de las probabilidades
numeros = [2, 3, 5]
probabilidades = [probabilidad_2, probabilidad_3, probabilidad_5]
colores = ['blue', 'green', 'red']

plt.bar(numeros, probabilidades, color=colores)
plt.xlabel('Número')
plt.ylabel('Probabilidad')
plt.title('Histograma de Probabilidades para los números 2, 3 y 5 en un dado equilibrado de 6 lados')
plt.xticks(numeros)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ventas por producto y día
ventas = np.array([
    [20, 15, 25, 30, 18, 22, 24],
    [12, 20, 14, 8, 15, 18, 16],
    [35, 28, 32, 30, 26, 24, 30],
    [40, 38, 45, 42, 39, 41, 37]
])

# media y mediana de ventas por producto
media_por_producto = np.mean(ventas, axis=1)
mediana_por_producto = np.median(ventas, axis=1)

# Plotear los datos de media y mediana por producto en un gráfico de barras
productos = ['A', 'B', 'C', 'D']
plt.bar(productos, media_por_producto, label='Media')
plt.bar(productos, mediana_por_producto, label='Mediana', alpha=0.7)
plt.xlabel('Producto')
plt.ylabel('Ventas')
plt.title('Media y Mediana de Ventas por Producto')
plt.legend()
plt.show()

# Calcular la media y mediana de ventas por día
media_por_dia = np.mean(ventas, axis=0)
mediana_por_dia = np.median(ventas, axis=0)

# Plotear los datos de media y mediana por día en un gráfico de trazo de líneas
dias = ['Lun', 'Mar', 'Mie', 'Jue', 'Vie', 'Sab', 'Dom']
plt.plot(dias, media_por_dia, label='Media')
plt.plot(dias, mediana_por_dia, label='Mediana', linestyle='dashed')
plt.xlabel('Día')
plt.ylabel('Ventas')
plt.title('Media y Mediana de Ventas por Día')
plt.legend()
plt.show()

# Conclusiones y estrategias
print("Conclusiones:")
print("- Las ventas promedio por producto y día varían. Jueves y Domingo hay un menor promedio de ventas")
print("\nEstrategias:")
print("- Realizar promociones especiales los días Jueves y Domingo.")
print("- Bajar los precios de los productos con menor demanda.")
print("- Ofrecer descuentos por la compra en combo de productos.")
print("- Armar un programa de fidelización y puntos para incentivar compras.")
print("- Aumentar la cantidad de productos más vendidos.")

# Calcular el desvío estándar por producto
desvio_estandar_por_producto = np.std(ventas, axis=1)

# Encontrar productos con mayor y menor desvío estándar
producto_mayor_desvio = productos[np.argmax(desvio_estandar_por_producto)]
producto_menor_desvio = productos[np.argmin(desvio_estandar_por_producto)]

# Plotear las ventas de los productos con mayor y menor desvío estándar
plt.plot(dias, ventas[productos.index(producto_mayor_desvio)], label=producto_mayor_desvio)
plt.plot(dias, ventas[productos.index(producto_menor_desvio)], label=producto_menor_desvio)
plt.xlabel('Día')
plt.ylabel('Ventas')
plt.title('Ventas de los Productos con Mayor y Menor Desvío Estándar')
plt.legend()
plt.show()
```

## EJERCICIO 5  

Solicitar al usuario que ingrese la cantidad N de datos a cargar. Luego generar de manera aleatoria los siguientes datos:

* SEXO= 'MASCULINO', 'FEMENINO','NO ESPECIFICA'
* EDAD= entre 28 y 60 años
* ALTURA= entre un rango de 165.0 mts hasta 210 mts
* PESO= entre un rango de 50  y 110 kg
* HORAS DE DESCANSO diario= un número decimal
* HORAS DE TRABAJO/ ESTUDIO diario= un número decimal
* HORAS DE EJERCICIOS diario= un número decimal

REalizar las siguientes indicaciones y extraer concluisiones sobre la información y la pooblación muestreada.

1.   Generar un array de dimensión 2 (matriz), con todos los datos ordenados, permitiendo visualizarlos como tabla.
2.   Suponiendo que la distribución de probabilidad de los datos muestrados de las N personas, se distribuyen en forma NORMAL. Encontrar la media en los datos de altura, peso, y horas de descanso, trabajo y ejercicios.
3.  Graficar la muestra o distribución de los datos  de pesos , y alturas en graficas separadas. Indicar el grado de variabilidad de sus datos
4.  Plotear de manera conjunta y encontrar si presentan correlación los datos de peso vs altura,
5. Plotear de manera conjunta los datos de horas de descanso, estudio y ejercicio



**Ayuda Como generar los datos**


```python
import numpy as np
import matplotlib.pyplot as plt

def generar_datos(N):
    sexo = np.random.choice(['MASCULINO', 'FEMENINO', 'NO ESPECIFICA'], N ,[0.4, 0.5, 0.1])  # Por ejemplo, 40% masculino, 50% femenino, 10% no especificado)
    # Generar edades con distribución normal
    edad =np.random.normal(44, 6, N)
    edad=np.clip(np.around(edad),28,60)
    # Generar alturas, pesos y otras características
    altura = np.random.normal(180, 10, N)
    altura = np.clip(np.around(altura, decimals=1), 150.0, 210.0)

    peso = np.random.normal(80, 15, N)
    peso = np.clip(np.around(peso, decimals=1), 40.0, 130.0)

    horas_trabajo_estudio = np.random.normal(10, 2, N)
    horas_trabajo_estudio = np.clip(np.around(horas_trabajo_estudio, decimals=1), 4.0, 12.0)

    horas_ejercicio = np.random.normal(1, 0.5, N)
    horas_ejercicio = np.clip(np.around(horas_ejercicio, decimals=1), 0, 2)

    horas_descanso = 24 - horas_trabajo_estudio - horas_ejercicio
    datos = np.column_stack((sexo, edad, altura, peso, horas_descanso, horas_trabajo_estudio, horas_ejercicio))
    return datos


# Solicitar al usuario la cantidad N de datos a cargar
N = int(input("Ingrese la cantidad de datos a cargar: "))

# Generar datos aleatorios para N personas con distribución normal y límites específicos
datos = generar_datos(N)

# Crear un array de dimensión 2 (matriz) con los datos
datos = np.array([edad, altura, peso, horas_descanso, horas_trabajo_estudio, horas_ejercicio])

# cambiar orden para visualizar los datos como tabla
datos_tabla = datos.T

# Calcular las medias de edad, altura, peso y horas
media_edad = np.mean(datos_tabla[:, 0])
media_altura = np.mean(datos_tabla[:, 1])
media_peso = np.mean(datos_tabla[:, 2])
media_horas_descanso = np.mean(datos_tabla[:, 3])
media_horas_trabajo_estudio = np.mean(datos_tabla[:, 4])
media_horas_ejercicio = np.mean(datos_tabla[:, 5])

# Graficar distribución de pesos
plt.hist(datos_tabla[:, 2], bins=20, edgecolor='black')
plt.xlabel('Peso')
plt.ylabel('Frecuencia')
plt.title('Distribución de Pesos')
plt.show()

# Graficar distribución de alturas
plt.hist(datos_tabla[:, 1], bins=20, edgecolor='black')
plt.xlabel('Altura')
plt.ylabel('Frecuencia')
plt.title('Distribución de Alturas')
plt.show()

# Graficar peso vs altura
plt.scatter(datos_tabla[:, 2], datos_tabla[:, 1])
plt.xlabel('Peso')
plt.ylabel('Altura')
plt.title('Peso vs Altura')
plt.show()

# Graficar horas de descanso, trabajo y ejercicio
plt.scatter(datos_tabla[:, 3], datos_tabla[:, 4], label='Descanso')
plt.scatter(datos_tabla[:, 3], datos_tabla[:, 5], label='Ejercicio')
plt.scatter(datos_tabla[:, 3], datos_tabla[:, 5], label='Trabajo/Estudio')
plt.xlabel('Horas de Descanso')
plt.ylabel('Horas')
plt.title('Horas de Descanso, Trabajo/Estudio y Ejercicio')
plt.legend()
plt.show()
```
