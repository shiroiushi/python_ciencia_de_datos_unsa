```python
#  A partir del dataset propuesto realiza una distribución de frecuencias de variables continuas

import matplotlib.pyplot as plt

# Obtener la columna de resultados
result_column = df['Result']

# Calcular la frecuencia de cada resultado
result_frequencies = result_column.value_counts()

# Crear un gráfico de barras para la distribución de frecuencias
result_frequencies.plot(kind='bar', color='green', edgecolor='black')
plt.title('Distribución de Resultados de Partidos')
plt.xlabel('Resultado')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45, fontsize=6)
plt.show()

# Imprimir las frecuencias de resultados
print("Frecuencias de Resultados:")
print(result_frequencies)
```
