# Queremos seleccionar las filas del DataFrame de pandas que corresponden a los goles de Lionel Messi, excepto los del año 2020.


Usamos el método loc para acceder a un grupo de filas y columnas por etiquetas.

El primer argumento es ‘Lionel Messi’, que indica que se seleccionan las filas que tienen ese valor como índice.

El segundo argumento es el método query, que permite hacer consultas sobre el DataFrame usando expresiones lógicas.

El argumento de query es ‘year != 2020’, que significa que se filtran las filas cuya columna ‘year’ sea distinta de 2020.

El resultado es un nuevo DataFrame con solo las filas que cumplen la condición.

```python
# Filtrando datos del dataset entre 2014 y 2019


messi_goals = df.loc[df['player_name'] == 'Lionel Messi'].query('year != 2020')
ronaldo_goals = df.loc[df['player_name'] =='Cristiano Ronaldo'].query('year != 2020')
messi_goals.head()
```

# Queremos crear un gráfico de líneas que muestra los goles de Lionel Messi y Cristiano Ronaldo por año.

El código hace lo siguiente:

Importa el módulo pyplot de matplotlib con el alias plt.

Crea una figura y unos ejes usando la función plt.subplots(), que devuelve un objeto Figure y un objeto Axes.

La figura es el contenedor donde se dibuja el gráfico, y los ejes son la región donde se representan los datos.

Dibuja dos líneas sobre los ejes usando el método ax.plot(), que recibe dos listas como argumentos: una con los valores del eje X (en este caso, los años) y otra con los valores del eje Y (en este caso, los goles).

Cada llamada a este método dibuja una línea con un color diferente por defecto.

Guarda el gráfico en un archivo llamado ‘1.jpg’ usando la función plt.savefig(), que recibe el nombre del archivo como argumento.

Muestra el gráfico en una ventana usando la función plt.show(), que no recibe ningún argumento.

El resultado del código es el siguiente gráfico:

```python
# Plot -> ax.plot(X, Y)
fig, ax = plt.subplots()
ax.plot(messi_goals.year, messi_goals.goals)
ax.plot(ronaldo_goals.year, ronaldo_goals.goals)
# plt.savefig('1.jpg')
plt.show()
```

# **Personalizar un plot**
Hay una serie de argumentos que te permitirán personalizar cómo se ven los datos de tu plot tanto como te puedas imaginar:

**marker='o'** → añade marcadores al plot y también puedes indicar el tipo de marcador que quieres.

Aquí te dejo todas las opciones de marcadores que hay:

**linestyle='--'** → cambia el tipo de línea.

 Así mismo, aquí podrás encontrar los distintos tipos de línea. Incluso si colocas 'None', simplemente ya no habrá líneas y quedarán solo los marcadores.

**color='b'** → cambia el color de tus datos. Aquí encontrarás cómo poner cualquier color e incluso dejarlos transparentes.

**label='name'** → pone una etiqueta al plot (muy útil cuando vas a poner varias gráficas en un solo plot).

Adicionalmente, el objeto Axes trae una serie de métodos que podrías usar para que sea sencillo identificar de qué trata tu gráfica. Colócalos siempre antes de mostrar tu gráfica con plt.show().

**.set_xlabel('Name')**→ coloca nombre al eje de las X.

**.set_ylabel('Name')**→ coloca nombre al eje de las Y.
**.set_title('Name')**→ coloca el título de tu plot.
**.set(xlabel='Name')**→ también puedes usar únicamente set y pasar como parámetros lo anterior.

**.legend()**→ si pusiste etiquetas, legend las mostrará.

```python
fig, ax = plt.subplots()
ax.plot(messi_goals.year, messi_goals.goals,
        marker='o', color='b', linestyle='--', label='Lionel Messi')
ax.plot(ronaldo_goals.year, ronaldo_goals.goals,
        marker='s', color='r', linestyle='dotted', label='Cristiano Ronaldo')

ax.set(xlabel='Years', ylabel='Goals', title='Messi vs. Cristiano goals (2014 - 2019)')
ax.legend()
# plt.savefig('2.jpg')
plt.show()
```

# **Cantidad de Tarjetas rojas**

Queremos crear un gráfico de barras que muestra las tarjetas rojas acumuladas por cada equipo. El código hace lo siguiente:

Agrupa el DataFrame df por la columna ‘team_name’ y suma los valores de la columna ‘red_cards’ para cada equipo usando el método groupby().sum().

Ordena los resultados de mayor a menor usando el método sort_values() con el argumento ascending=False.

Selecciona los primeros 10 equipos con más tarjetas rojas usando el método head(10).
Guarda el resultado en una variable llamada top_red_cards.

Crea un gráfico de barras con la función plt.bar(), pasando como argumentos el índice de top_red_cards (los nombres de los equipos) y los valores de top_red_cards (las tarjetas rojas).

Rota las etiquetas del eje X 90 grados usando la función plt.xticks() con el argumento rotation=90, para que se vean mejor.

Pone un título al gráfico con la función plt.title(), pasando como argumento una cadena de texto.

Pone un nombre al eje Y con la función plt.ylabel(), pasando como argumento una cadena de texto.

Muestra el gráfico en una ventana usando la función plt.show().

El resultado del código es el siguiente gráfico:

```python
# Datos de tarjetas rojas por equipo
top_red_cards = df.groupby('team_name')['red_cards'].sum().sort_values(ascending=False).head(10)

# Plot
# plt.figure(figsize=[6.4, 6])
plt.style.use('ggplot')
plt.bar(top_red_cards.index, top_red_cards)
plt.xticks(rotation=90)
plt.title('Tarjetas rojas acumuladas por equipo')
plt.ylabel('Red cards')
plt.show()
```

```python
# Datos de tarjetas rojas por jugador
top_red_cards = df.groupby('player_name')['red_cards'].sum().sort_values(ascending=False).head(10)

# Plot
# plt.figure(figsize=[6.4, 6])
plt.style.use('ggplot')
plt.bar(top_red_cards.index, top_red_cards)
plt.xticks(rotation=90)
plt.title('Tarjetas rojas acumuladas por jugador')
plt.ylabel('Red cards')
plt.show()
```

Por ejemplo, si quieres saber el número total de goles por equipo, puedes usar groupby(‘team_name’).sum() sobre la columna ‘goals’. Esto te devolverá un nuevo DataFrame con el nombre de cada equipo como índice y la suma de sus goles como valor.

```python
top_red_goals=df.groupby("team_name")["goals"].sum().sort_values(ascending=False).head(10)

plt.style.use('ggplot')
plt.bar(top_red_goals.index, top_red_goals)
plt.xticks(rotation=90)
plt.title('Goles por equipo')
plt.ylabel('Goles')
plt.show()
```

# **Gráfico de barras apilado**

```python
# Filter
df['goals_assists'] = df.goals + df.assists
top_players = df.groupby('player_name')[['goals', 'assists', 'goals_assists']].sum() \
                .sort_values('goals_assists', ascending=False).head(10)

# Plot
plt.bar(top_players.index, top_players.goals, label='Goals')
plt.bar(top_players.index, top_players.assists, label='Assists', bottom=top_players.goals)
plt.xticks(rotation=90)
plt.title('Goles y asistencias de los top players')
plt.ylabel('Goals and assists')
plt.legend()
plt.show()
```

# **Histograma**
Para graficar un histograma se usa plt.hist(X) y como argumento se pasa la columna que quieras analizar. Además, puedes personalizar el gráfico con otros argumentos:

bins=n → dice cuántas barras serán mostradas en el histograma. Por defecto vienen 10.

A este argumento también le puedes pasar una lista para establecer un límite en el rango de las X en las que quieras que se muestren.

Si quieres colocar 2 histogramas que se podrían sobreponer entre sí, se puede usar el argumento alpha que si le pasas un valor entre 0 y 1 cambia la transparencia del color.

También puedes usar el argumento histtype='step' que hace que cambie la forma que se representa el histograma, por ejemplo líneas en vez de rectángulos sólidos. Aquí te dejo la documentación para ver qué otros tipos de histtype existen y también otros argumentos de hist.

Por último, si usas density=True, el gráfico se trasformará a porcentajes para que el área total de los rectángulos sea de 1 (100%). Te recomiendo recordar este argumento porque podría ser muy útil más adelante para entender el peso de los valores más que su cantidad en sí misma.

```python
# Goles totales por equipo en 2018
df_2018 = df.query('year == 2018')
team_goals_2018 = df_2018.groupby('team_name')['goals'].sum()

# Plot
plt.figure(figsize=[7, 6]) # 7 in de ancho y 6 in de alto
plt.hist(team_goals_2018, bins=25)
plt.title('Goles totales por equipo en 2018')
plt.xlabel('Cantidad de goles')
plt.ylabel('Número de equipos con X cantidad de goles')
plt.savefig('6.jpg')
plt.show()
```

```python
# Goles totales por equipo en 2018
df_2018 = df.query('year == 2018')
team_goals_2018 = df_2018.groupby('team_name')['goals'].sum()
plt.hist(team_goals_2018, bins=15, histtype='step', density=True)
plt.show()
```

# **Gráfico de dispersión o scatter plot**
Se lo plotea con plt.scatter(x, y). Se le puede colocar varios argumentos que ya vimos para los gráficos anteriores, pero adicionalmente, hay un par muy útiles para pasar una tercera variable, por ejemplo, de tiempo. Con c=col podrás colorear cada punto con un degradado. Y con s=col podrás darles distintos tamaños a los puntos. Si quieres ver otros argumentos de este tipo de gráfico, acá te dejo la documentación oficial.

```python
plt.scatter(df_2018.shots, df_2018.goals, c=df_2018.games)
plt.title('Disparos vs. goles')
plt.xlabel('Disparos')
plt.ylabel('Goles')
plt.savefig('7.jpg')
plt.show()
```
