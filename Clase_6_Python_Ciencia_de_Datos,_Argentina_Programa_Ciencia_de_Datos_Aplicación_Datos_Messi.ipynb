# Queremos seleccionar las filas del DataFrame de pandas que corresponden a los goles de Lionel Messi, excepto los del ao 2020.


Usamos el mtodo loc para acceder a un grupo de filas y columnas por etiquetas.

El primer argumento es Lionel Messi, que indica que se seleccionan las filas que tienen ese valor como ndice.

El segundo argumento es el mtodo query, que permite hacer consultas sobre el DataFrame usando expresiones lgicas.

El argumento de query es year != 2020, que significa que se filtran las filas cuya columna year sea distinta de 2020.

El resultado es un nuevo DataFrame con solo las filas que cumplen la condicin.

```python
# Filtrando datos del dataset entre 2014 y 2019


messi_goals = df.loc[df['player_name'] == 'Lionel Messi'].query('year != 2020')
ronaldo_goals = df.loc[df['player_name'] =='Cristiano Ronaldo'].query('year != 2020')
messi_goals.head()
```

# Queremos crear un grfico de lneas que muestra los goles de Lionel Messi y Cristiano Ronaldo por ao.

El cdigo hace lo siguiente:

Importa el mdulo pyplot de matplotlib con el alias plt.

Crea una figura y unos ejes usando la funcin plt.subplots(), que devuelve un objeto Figure y un objeto Axes.

La figura es el contenedor donde se dibuja el grfico, y los ejes son la regin donde se representan los datos.

Dibuja dos lneas sobre los ejes usando el mtodo ax.plot(), que recibe dos listas como argumentos: una con los valores del eje X (en este caso, los aos) y otra con los valores del eje Y (en este caso, los goles).

Cada llamada a este mtodo dibuja una lnea con un color diferente por defecto.

Guarda el grfico en un archivo llamado 1.jpg usando la funcin plt.savefig(), que recibe el nombre del archivo como argumento.

Muestra el grfico en una ventana usando la funcin plt.show(), que no recibe ningn argumento.

El resultado del cdigo es el siguiente grfico:

```python
# Plot -> ax.plot(X, Y)
fig, ax = plt.subplots()
ax.plot(messi_goals.year, messi_goals.goals)
ax.plot(ronaldo_goals.year, ronaldo_goals.goals)
# plt.savefig('1.jpg')
plt.show()
```

# **Personalizar un plot**
Hay una serie de argumentos que te permitirn personalizar cmo se ven los datos de tu plot tanto como te puedas imaginar:

**marker='o'**  aade marcadores al plot y tambin puedes indicar el tipo de marcador que quieres.

Aqu te dejo todas las opciones de marcadores que hay:

**linestyle='--'**  cambia el tipo de lnea.

 As mismo, aqu podrs encontrar los distintos tipos de lnea. Incluso si colocas 'None', simplemente ya no habr lneas y quedarn solo los marcadores.

**color='b'**  cambia el color de tus datos. Aqu encontrars cmo poner cualquier color e incluso dejarlos transparentes.

**label='name'**  pone una etiqueta al plot (muy til cuando vas a poner varias grficas en un solo plot).

Adicionalmente, el objeto Axes trae una serie de mtodos que podras usar para que sea sencillo identificar de qu trata tu grfica. Colcalos siempre antes de mostrar tu grfica con plt.show().

**.set_xlabel('Name')** coloca nombre al eje de las X.

**.set_ylabel('Name')** coloca nombre al eje de las Y.
**.set_title('Name')** coloca el ttulo de tu plot.
**.set(xlabel='Name')** tambin puedes usar nicamente set y pasar como parmetros lo anterior.

**.legend()** si pusiste etiquetas, legend las mostrar.

```python
fig, ax = plt.subplots()
ax.plot(messi_goals.year, messi_goals.goals,
        marker='o', color='b', linestyle='--', label='Lionel Messi')
ax.plot(ronaldo_goals.year, ronaldo_goals.goals,
        marker='s', color='r', linestyle='dotted', label='Cristiano Ronaldo')

ax.set(xlabel='Years', ylabel='Goals', title='Messi vs. Cristiano goals (2014 - 2019)')
ax.legend()
# plt.savefig('2.jpg')
plt.show()
```

# **Cantidad de Tarjetas rojas**

Queremos crear un grfico de barras que muestra las tarjetas rojas acumuladas por cada equipo. El cdigo hace lo siguiente:

Agrupa el DataFrame df por la columna team_name y suma los valores de la columna red_cards para cada equipo usando el mtodo groupby().sum().

Ordena los resultados de mayor a menor usando el mtodo sort_values() con el argumento ascending=False.

Selecciona los primeros 10 equipos con ms tarjetas rojas usando el mtodo head(10).
Guarda el resultado en una variable llamada top_red_cards.

Crea un grfico de barras con la funcin plt.bar(), pasando como argumentos el ndice de top_red_cards (los nombres de los equipos) y los valores de top_red_cards (las tarjetas rojas).

Rota las etiquetas del eje X 90 grados usando la funcin plt.xticks() con el argumento rotation=90, para que se vean mejor.

Pone un ttulo al grfico con la funcin plt.title(), pasando como argumento una cadena de texto.

Pone un nombre al eje Y con la funcin plt.ylabel(), pasando como argumento una cadena de texto.

Muestra el grfico en una ventana usando la funcin plt.show().

El resultado del cdigo es el siguiente grfico:

```python
# Datos de tarjetas rojas por equipo
top_red_cards = df.groupby('team_name')['red_cards'].sum().sort_values(ascending=False).head(10)

# Plot
# plt.figure(figsize=[6.4, 6])
plt.style.use('ggplot')
plt.bar(top_red_cards.index, top_red_cards)
plt.xticks(rotation=90)
plt.title('Tarjetas rojas acumuladas por equipo')
plt.ylabel('Red cards')
plt.show()
```

```python
# Datos de tarjetas rojas por jugador
top_red_cards = df.groupby('player_name')['red_cards'].sum().sort_values(ascending=False).head(10)

# Plot
# plt.figure(figsize=[6.4, 6])
plt.style.use('ggplot')
plt.bar(top_red_cards.index, top_red_cards)
plt.xticks(rotation=90)
plt.title('Tarjetas rojas acumuladas por jugador')
plt.ylabel('Red cards')
plt.show()
```

Por ejemplo, si quieres saber el nmero total de goles por equipo, puedes usar groupby(team_name).sum() sobre la columna goals. Esto te devolver un nuevo DataFrame con el nombre de cada equipo como ndice y la suma de sus goles como valor.

```python
top_red_goals=df.groupby("team_name")["goals"].sum().sort_values(ascending=False).head(10)

plt.style.use('ggplot')
plt.bar(top_red_goals.index, top_red_goals)
plt.xticks(rotation=90)
plt.title('Goles por equipo')
plt.ylabel('Goles')
plt.show()
```

# **Grfico de barras apilado**

```python
# Filter
df['goals_assists'] = df.goals + df.assists
top_players = df.groupby('player_name')[['goals', 'assists', 'goals_assists']].sum() \
                .sort_values('goals_assists', ascending=False).head(10)

# Plot
plt.bar(top_players.index, top_players.goals, label='Goals')
plt.bar(top_players.index, top_players.assists, label='Assists', bottom=top_players.goals)
plt.xticks(rotation=90)
plt.title('Goles y asistencias de los top players')
plt.ylabel('Goals and assists')
plt.legend()
plt.show()
```

# **Histograma**
Para graficar un histograma se usa plt.hist(X) y como argumento se pasa la columna que quieras analizar. Adems, puedes personalizar el grfico con otros argumentos:

bins=n  dice cuntas barras sern mostradas en el histograma. Por defecto vienen 10.

A este argumento tambin le puedes pasar una lista para establecer un lmite en el rango de las X en las que quieras que se muestren.

Si quieres colocar 2 histogramas que se podran sobreponer entre s, se puede usar el argumento alpha que si le pasas un valor entre 0 y 1 cambia la transparencia del color.

Tambin puedes usar el argumento histtype='step' que hace que cambie la forma que se representa el histograma, por ejemplo lneas en vez de rectngulos slidos. Aqu te dejo la documentacin para ver qu otros tipos de histtype existen y tambin otros argumentos de hist.

Por ltimo, si usas density=True, el grfico se trasformar a porcentajes para que el rea total de los rectngulos sea de 1 (100%). Te recomiendo recordar este argumento porque podra ser muy til ms adelante para entender el peso de los valores ms que su cantidad en s misma.

```python
# Goles totales por equipo en 2018
df_2018 = df.query('year == 2018')
team_goals_2018 = df_2018.groupby('team_name')['goals'].sum()

# Plot
plt.figure(figsize=[7, 6]) # 7 in de ancho y 6 in de alto
plt.hist(team_goals_2018, bins=25)
plt.title('Goles totales por equipo en 2018')
plt.xlabel('Cantidad de goles')
plt.ylabel('Nmero de equipos con X cantidad de goles')
plt.savefig('6.jpg')
plt.show()
```

```python
# Goles totales por equipo en 2018
df_2018 = df.query('year == 2018')
team_goals_2018 = df_2018.groupby('team_name')['goals'].sum()
plt.hist(team_goals_2018, bins=15, histtype='step', density=True)
plt.show()
```
